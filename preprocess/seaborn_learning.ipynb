```python
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('tips.csv')
```

```python
dataset.head()
```

```python
# 1. 可视化关系relplot
# 1.1 scatter图
sns.relplot(data = dataset, x = 'total_bill', y = 'tip', hue = 'size')
```

```python
dowjones = pd.read_csv("dowjones.csv")
dowjones.head()
```

```python
# 1.2 line图
sns.relplot(data = dowjones, x = 'Date', y = 'Price', kind = 'line')
```

```python
fmri = pd.read_csv("fmri.csv")
fmri.head()
```

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', kind = 'line')
```

```python
# 2. 可视化分布displot
# 2.1 hist图
penguins = pd.read_csv("penguins.csv")
penguins.head()
```

```python
sns.displot(penguins, x = 'flipper_length_mm', bins = 20, kde = True)
```

```python
tips = pd.read_csv("tips.csv")
sns.displot(tips, x="size")
```

```python
sns.displot(tips, x="size", bins=[1, 2, 3, 4, 5, 6, 7])
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="sex", stat = 'density', common_norm = False)
```

```python
# 2.2 kde图
sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=.25)
```

```python
# 2.3 ecdf图， 经验累积分布图
sns.displot(penguins, x="flipper_length_mm", kind="ecdf")
```

```python
# 可视化分布的二元作图
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", cbar=True)
```

```python
# 2.4 联合和边缘分布图
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
# 2.5 pair分布
sns.pairplot(penguins)
```

```python
# 3. 可视化分类
# 3.1 分类散点图
# 3.1.1 strip图
tips = pd.read_csv("tips.csv")
sns.catplot(data=tips, x="day", y="total_bill")  # jitter = False 可以防止左右抖动
```

```python
# 3.1.2 swarm图
sns.catplot(data=tips, x="day", y="total_bill", hue = 'sex', kind="swarm")
```

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="swarm")
```

```python
# 3.2 比较分布图
# 3.2.1 box图
sns.catplot(data=tips, x="day", y="total_bill", kind="box", hue="smoker")
```

```python
diamonds = pd.read_csv("diamonds.csv")
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color", y="price", kind="boxen",
)
```

```python
# 3.2.2 violin图
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex", kind="violin",
)
```

```python
# 3.3 分类估计图
# 3.3.1 bar图
titanic = pd.read_csv("titanic.csv")
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar")
```

```python
sns.catplot(data=titanic, x="age", y="deck", errorbar=("pi", 95), kind="bar")
```

```python
# 3.3.2 count图
sns.catplot(data=titanic, x="deck", kind="count")
```

```python
# 3.3.3 point图
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="point")
```
